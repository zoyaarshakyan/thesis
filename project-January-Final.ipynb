```python
# importing literature
import numpy as np

from pandas.tseries import *
from pandas_datareader.data import DataReader
from pandas_datareader import data
import pandas as pd

import statsmodels.api as sm

import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.stats.stattools import durbin_watson

# importing monthly returns of assets from May 2000 to December 2017
data = pd.read_excel('forJan.xlsx')
data.set_index(data['Date'], inplace=True)
data.set_index('Date', inplace=True)
```

```python
#  testing stationarity of returns 
from arch.unitroot import ADF
for column in data:
    adf = ADF(data[column])
    print(adf.summary().as_text())
```

```python
for column in data:
    mod = sm.tsa.MarkovAutoregression(data[column], k_regimes=2, order=1, trend='c', switching_ar=True, switching_variance=True)
    res = mod.fit(cov_type="opg", maxiter=100, full_output=True,disp=True,em_iter=1)
    print (res.summary())
    print  res.params
    print  res.smoothed_marginal_probabilities[1].tail(1)
    #In-sample prediction and out-of-sample forecasting
    predict= mod.predict(res.params, start=None, end=None, probabilities=None, conditional=False)
    print res.resid[-1:]
    print predict[-1]
    # testing autocorrelation in errors
    print durbin_watson(res.resid)
```
